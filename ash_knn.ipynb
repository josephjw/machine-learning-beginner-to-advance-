```python
#converting categorical values of 'Gender' to numerical format
data['Class']=data['Class'].apply({'p':'1','e':'2'}.get)
# .cap-shape: bell=b,conical=c,convex=x,flat=f, knobbed=k,sunken=s
data['Cap-shape']=data['Cap-shape'].apply({'b':'1','c':'2','x':'3','f':'4','k':'5','s':'6'}.get)
# 2. cap-surface: fibrous=f,grooves=g,scaly=y,smooth=s
data['Cap-surface']=data['Cap-surface'].apply({'f':'1','g':'2','y':'3','s':'4'}.get)
# 3. cap-color: brown=n,buff=b,cinnamon=c,gray=g,green=r, pink=p,purple=u,red=e,white=w,yellow=y
data['Cap-color']=data['Cap-color'].apply({'n':'1','b':'2','c':'3','g':'4','p':'5','u':'6','e':'7','w':'8','y':'9'}.get)
# 4. bruises?: bruises=t,no=f
data['Bruises']=data['Bruises'].apply({'t':'1','f':'2'}.get)
# 5. odor: almond=a,anise=l,creosote=c,fishy=y,foul=f, musty=m,none=n,pungent=p,spicy=s
data['Odor']=data['Class'].apply({'a':'1','l':'2','c':'3','y':'4','f':'5','m':'6','n':'7','p':'8','s':'9'}.get)
# 6. gill-attachment: attached=a,descending=d,free=f,notched=n
data['Gill-attachment']=data['Gill-attachment'].apply({'a':'1','d':'2','f':'3','n':'4'}.get)
# 7. gill-spacing: close=c,crowded=w,distant=d
data['Gill-spacing']=data['Gill-spacing'].apply({'c':'1','e':'2','d':'3'}.get)
# 8. gill-size: broad=b,narrow=n
data['Gill-size']=data['Gill-size'].apply({'p':'1','n':'2'}.get)
# 9. gill-color: black=k,brown=n,buff=b,chocolate=h,gray=g, green=r,orange=o,pink=p,purple=u,red=e,
# white=w,yellow=y
data['Gill-color']=data['Gill-color'].apply({'k':'1','n':'2','b':'3','h':'4','g':'5','r':'6','o':'7','p':'8','u':'9','e':'10'}.get)
# 10. stalk-shape: enlarging=e,tapering=t
data['Stalk-shape']=data['Class'].apply({'e':'1','t':'2'}.get)
# 11. stalk-root: bulbous=b,club=c,cup=u,equal=e, rhizomorphs=z,rooted=r,missing=?
data['Stalk-root']=data['Class'].apply({'b':'1','c':'2','u':'3','e':'4','z':'5','r':'6','?':'7'}.get)
# 12. stalk-surface-above-ring: fibrous=f,scaly=y,silky=k,smooth=s
data['Stalk-surface-above-ring']=data['Stalk-surface-above-ring'].apply({'f':'1','y':'2','k':'3','s':'4'}.get)
# 13. stalk-surface-below-ring: fibrous=f,scaly=y,silky=k,smooth=s
data['Stalk-surface-below-ring']=data['Stalk-surface-below-ring'].apply({'f':'1','y':'2','k':'3','s':'4'}.get)
# 14. stalk-color-above-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o, pink=p,red=e,white=w,yellow=y
data['Stalk-color-above-ring']=data['Stalk-color-above-ring'].apply({'n':'1','b':'2','c':'3','g':'4','o':'5','p':'6','w':'7','e':'8','y':'9'}.get)
# 15. stalk-color-below-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o, pink=p,red=e,white=w,yellow=y
data['Stalk-color-below-ring']=data['Stalk-color-below-ring'].apply({'n':'1','b':'2','c':'3','g':'4','o':'5','p':'6','w':'7','e':'8','y':'9'}.get)
# 16. veil-type: partial=p,universal=u
data['Veil-type']=data['Veil-type'].apply({'p':'1','u':'2'}.get)
# 17. veil-color: brown=n,orange=o,white=w,yellow=y
data['Veil-color']=data['Veil-color'].apply({'n':'1','o':'2','w':'3','y':'4'}.get)
# 18. ring-number: none=n,one=o,two=t
data['Ring-number']=data['Ring-number'].apply({'n':'1','o':'2','t':'3'}.get)
# 19. ring-type: cobwebby=c,evanescent=e,flaring=f,large=l, none=n,pendant=p,sheathing=s,zone=z
data['Ring-type']=data['Ring-type'].apply({'c':'1','e':'2','f':'3','l':'4','n':'5','p':'6','s':'7','z':'8'}.get)
# 20. spore-print-color: black=k,brown=n,buff=b,chocolate=h,green=r, orange=o,purple=u,white=w,yellow=y
data['Spore-print-color']=data['Spore-print-color'].apply({'k':'1','n':'2','b':'3','h':'4','r':'5','o':'6','u':'7','w':'8','y':'9'}.get)
# 21. population: abundant=a,clustered=c,numerous=n, scattered=s,several=v,solitary=y
data['Population']=data['Population'].apply({'a':'1','c':'2','n':'3','s':'4','s':'5','y':'6'}.get)
# 22. habitat: grasses=g,leaves=l,meadows=m,paths=p, urban=u,waste=w,woods=d
data['Habitat']=data['Habitat'].apply({'p':'1','u':'2','w':'3','d':'4','l':'5','g':'6','m':'7'}.get)

```

```python
#divide the dataset into dependent and independent columns
data=data.fillna(0)
data=data.fillna(value=np.nan)
x=data.loc[:,data.columns!= 'Habitat']#independent in nature
y=data['Habitat']#dependent column
```

```python
#split the data into training and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#importing KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3)
```

```python
classifier.fit(x_train,y_train)
```

```python
predictions=classifier.predict(x_test)
```

```python
#accuracy computation
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,predictions)
ac=accuracy_score(y_test,predictions)
```
